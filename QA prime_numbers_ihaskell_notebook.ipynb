# Prime Numbers Generator using Sieve of Eratosthenes

This notebook generates a list of prime numbers using the Sieve of Eratosthenes algorithm.

```haskell
-- Prime numbers generator using the Sieve of Eratosthenes
primes :: [Int]
primes = sieve [2..]
  where
    sieve (p:xs) = p : sieve [x | x <- xs, x `mod` p /= 0]

-- Display the first 50 prime numbers
take 50 primes
```
